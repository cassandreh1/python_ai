### Train test split data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8) #80 % of data for training purpose 20% will be used for testing 
```

```python
# MI untrained model and will learn from the training data
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
# create object of a classes
lnr = LinearRegression()
rdf = RandomForestRegressor()
```

```python
lnr.fit(x_train, y_train) # training purpose
```

```python
rdf.fit(x_train, y_train)
```

```python
#perfromance
print(lnr.score(x_train, y_train) * 100)
print("RandomForest Regression: ", rdf.score(x_train, y_train))
```
